```python
import json
import pandas as pd
import time
import datetime
```

```python
ITERATIONS = 16
wallet_address = 'bc1q4fdqv7z8mqk73rtgzk7qdxnzw675320gjar6xx'
```

```python
i = 0
in_transactions_data = []
count = 0
while i <= ITERATIONS:
    f_name = f'tx{i*100}.json'
    print(f'opeing file {f_name}')
    file = open(f_name, 'r')
    txs = json.load(file)
    address = 'bc1q4fdqv7z8mqk73rtgzk7qdxnzw675320gjar6xx'
    
    
    for index in txs:
        addresses = set()
        tx = txs[index]
        tx_hash = tx['hash']
        tx_time = datetime.datetime.fromtimestamp(tx['time'])
        inputs = tx['inputs']
        outs = tx['out']
        fee = tx['fee']
        is_out_tx = False
        tx_amount = 0
        count += 1
        for input in inputs:
            prev = input['prev_out']
            if prev['addr'] == address:
                is_out_tx= True
            else:
                addresses.add(prev['addr'])

        if not is_out_tx:
            for out_tx in outs:
                if out_tx['addr'] == address:
                    tx_amount = float(out_tx['value']) * 0.00000001
                    out_tx_address = out_tx['addr']
                    in_transactions_data.append([tx_hash, tx_time, '##'.join(str(s) for s in addresses), tx_amount, out_tx_address])
    i = i + 1
print(count)
```

```python
result = pd.DataFrame(in_transactions_data, columns=['tx_hash', 'tx_date_time', 'from_addr', 'amount', 'to_addr'])
result
```
